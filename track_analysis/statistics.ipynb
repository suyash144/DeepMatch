```python
for key in mouse_dicts.keys():
    # unique_unitmatch, unique_dnn, unique_func, overlap_unitmatch_dnn, overlap_unitmatch_func, overlap_dnn_func, overlap_all
    AllDNN = mouse_dicts[key][1] + mouse_dicts[key][3] + mouse_dicts[key][5] + mouse_dicts[key][6]
    AllUM = mouse_dicts[key][0] + mouse_dicts[key][3] + mouse_dicts[key][4] + mouse_dicts[key][6]
    AllFunc = mouse_dicts[key][2] + mouse_dicts[key][4] + mouse_dicts[key][5] + mouse_dicts[key][6]
    AllPred = mouse_dicts[key][0] + mouse_dicts[key][1] + mouse_dicts[key][2] + mouse_dicts[key][3] + mouse_dicts[key][4] + mouse_dicts[key][5] + mouse_dicts[key][6]
    uniDNN_over_AllDNN.append(mouse_dicts[key][1]/AllDNN)
    uniUM_over_AllUM.append(mouse_dicts[key][0]/AllUM)
    uniFunc_over_AllFunc.append(mouse_dicts[key][2]/AllFunc)
    DNNFunc_over_AllDNN.append(mouse_dicts[key][5]/AllDNN)
    UMFunc_over_AllUM.append(mouse_dicts[key][4]/AllUM)
    DNNFunc_over_AllDNN2.append((mouse_dicts[key][5]+mouse_dicts[key][6])/AllDNN)
    UMFunc_over_AllUM2.append((mouse_dicts[key][4]+mouse_dicts[key][6])/AllUM)
    DNNFunc_over_AllPred.append(mouse_dicts[key][5]/AllPred)
    UMFunc_over_AllPred.append(mouse_dicts[key][4]/AllPred)
    DNNUM_over_AllPred.append(mouse_dicts[key][3]/AllPred)
    DNNFunc_over_AllFunc.append((mouse_dicts[key][5]+mouse_dicts[key][6])/AllFunc)
    UMFunc_over_AllFunc.append((mouse_dicts[key][4]+mouse_dicts[key][6])/AllFunc)

# change the list to numpy array
uniDNN_over_AllDNN = np.array(uniDNN_over_AllDNN)
uniUM_over_AllUM = np.array(uniUM_over_AllUM)
uniFunc_over_AllFunc = np.array(uniFunc_over_AllFunc)
DNNFunc_over_AllDNN = np.array(DNNFunc_over_AllDNN)
UMFunc_over_AllUM = np.array(UMFunc_over_AllUM)
DNNFunc_over_AllDNN2 = np.array(DNNFunc_over_AllDNN2)
UMFunc_over_AllUM2 = np.array(UMFunc_over_AllUM2)
DNNFunc_over_AllPred = np.array(DNNFunc_over_AllPred)
UMFunc_over_AllPred = np.array(UMFunc_over_AllPred)
DNNUM_over_AllPred = np.array(DNNUM_over_AllPred)
DNNFunc_over_AllFunc = np.array(DNNFunc_over_AllFunc)
UMFunc_over_AllFunc = np.array(UMFunc_over_AllFunc)

# remove the nan values
uniDNN_over_AllDNN = uniDNN_over_AllDNN[~np.isnan(uniDNN_over_AllDNN)]
uniUM_over_AllUM = uniUM_over_AllUM[~np.isnan(uniUM_over_AllUM)]
uniFunc_over_AllFunc = uniFunc_over_AllFunc[~np.isnan(uniFunc_over_AllFunc)]
DNNFunc_over_AllDNN = DNNFunc_over_AllDNN[~np.isnan(DNNFunc_over_AllDNN)]
UMFunc_over_AllUM = UMFunc_over_AllUM[~np.isnan(UMFunc_over_AllUM)]
DNNFunc_over_AllDNN2 = DNNFunc_over_AllDNN2[~np.isnan(DNNFunc_over_AllDNN2)]
UMFunc_over_AllUM2 = UMFunc_over_AllUM2[~np.isnan(UMFunc_over_AllUM2)]
DNNFunc_over_AllPred = DNNFunc_over_AllPred[~np.isnan(DNNFunc_over_AllPred)]
UMFunc_over_AllPred = UMFunc_over_AllPred[~np.isnan(UMFunc_over_AllPred)]
DNNUM_over_AllPred = DNNUM_over_AllPred[~np.isnan(DNNUM_over_AllPred)]
DNNFunc_over_AllFunc = DNNFunc_over_AllFunc[~np.isnan(DNNFunc_over_AllFunc)]
UMFunc_over_AllFunc = UMFunc_over_AllFunc[~np.isnan(UMFunc_over_AllFunc)]
```

```python
### statistics for unique/AllMethod ratio of the methods
# Perform independent t-tests
ttest_1_2 = stats.ttest_ind(uniDNN_over_AllDNN, uniUM_over_AllUM, equal_var=False)
ttest_1_3 = stats.ttest_ind(uniDNN_over_AllDNN, uniFunc_over_AllFunc, equal_var=False)
ttest_2_3 = stats.ttest_ind(uniUM_over_AllUM, uniFunc_over_AllFunc, equal_var=False)

# Create bar plot with error bars
means = [np.mean(uniDNN_over_AllDNN), np.mean(uniUM_over_AllUM), np.mean(uniFunc_over_AllFunc)]
# stds = [np.std(uniDNN_over_AllDNN), np.std(uniUM_over_AllUM), np.std(uniFunc_over_AllFunc)]
sems = [stats.sem(uniDNN_over_AllDNN), stats.sem(uniUM_over_AllUM), stats.sem(uniFunc_over_AllFunc)]
labels = ['1', '2', '3']
def add_significance_text(ax, x1, x2, y, h, p_value, offset_index):
    """
    Add a line and significance label ('ns' for not significant, '*' for significant) 
    between two bars at staggered heights.
    
    Parameters:
    ax: The axis object to add the annotation to.
    x1, x2: The x positions of the bars to compare.
    y: The y position of the line end.
    h: The height of the lines connecting the bars.
    p_value: The p-value used to determine if a star should be added.
    offset_index: The index to determine the height offset for staggered lines.
    """
    # Stagger the heights of the lines
    stagger = 0.05 * offset_index
    ax.plot([x1, x1, x2, x2], [y+h+stagger, y+stagger, y+stagger, y+h+stagger], lw=0.8, c='k')
    # Based on the p-value, choose the label
    # label = '*' if p_value < 0.05 else 'ns'
    # if p_value >= 0.05:
    #     label = 'ns'
    # elif p_value<0.05 and p_value>=0.01:
    #     label = '*'
    # elif p_value<0.01 and p_value>=0.001:
    #     label = '**'
    # else:
    #     label = '***'
    # Output to the fourth decimal place
    p_value = round(p_value, 4)
    label = 'p = ' + str(p_value)
    # labelname = 'p = ' + str(p_value)
    # label = labelname if p_value < 0.05 else 'ns'
    ax.text((x1+x2)*0.5, y+h+stagger, label, ha='center', va='bottom', color='k')

# Create a new plot with staggered significance lines
fig, ax = plt.subplots(figsize=(6, 6))

# Bar plot
bars = ax.bar(labels, means, yerr=sems, capsize=10, color=['blue', 'orange', 'grey'], alpha=0.5)

# Calculate the y position for the significance line, which will be above the error bars
y_offsets = [sem * 1.5 for sem in sems]  # This will be the distance above the error bars
y_values = [mean + offset for mean, offset in zip(means, y_offsets)]

# The maximum y value for drawing the significance lines
max_y = max(y_values) + 0.02

# Add staggered significance lines
# The offset_index parameter will stagger the heights of the significance lines
add_significance_text(ax, 0, 1, max_y, 0.01, ttest_1_2.pvalue, 1)
add_significance_text(ax, 1, 2, max_y, 0.01, ttest_2_3.pvalue, 2)
add_significance_text(ax, 0, 2, max_y, 0.01, ttest_1_3.pvalue, 3)

# Set labels and title
ax.set_ylabel('Unique/AllMethod ratio', fontsize=16)
ax.set_title('Unique/AllMethod predicted pairs ratio', fontsize=18)
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(['DNN', 'UM', 'Func'])
ax.tick_params(axis='x', labelsize=16)
plt.show()
```
